월별 액티브한 사용자들 수 카운트.

```
%%sql

select to_char(DATE(st.ts), 'YYYY-MM') as yearmonth, count(distinct usc.userid)
FROM raw_data.session_timestamp st
JOIN raw_data.user_session_channel usc
ON st.sessionid = usc.sessionid
GROUP BY yearmonth
ORDER BY yearmonth
```
